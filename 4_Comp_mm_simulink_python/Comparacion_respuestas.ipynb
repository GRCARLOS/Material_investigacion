# Comparamos las respuestas obtenidas para un loop de control en corriente del solver RKDP en python  respecto  de la respuesta ofrecidad por Simulink

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

## import the simulink data file
sm=pd.read_csv('DATO_150823.csv')  

## import the python generated data file
py= pd.read_csv('DATO_python_150823.csv')
```

```python
#simulink data
ts=np.array(sm.Tiempo)
Is=np.array(sm.Corriente)
Vs=np.array(sm.Velocidad)

#python data
tp=np.array(py.Tiempo)
Ip=np.array(py.Corriente)
Vp=np.array(py.Velocidad)
#grafica completa
plt.figure(figsize=(8, 6)) 
plt.plot(ts, Is, label='Simulink')
plt.plot(tp, Ip, label='Python')
plt.legend()
plt.title('Respuesta de corriente')
plt.grid(True)
plt.xlabel('Tiempo (s)')
plt.ylabel('Corriente (A) ')
plt.show()
#Grafica acotada a 10 milisegundos
plt.figure(figsize=(8, 6)) 
plt.plot(ts[0:10], Is[0:10], label='Simulink')
plt.plot(tp[0:10], Ip[0:10], label='Python')
plt.legend()
plt.title('Respuesta de corriente')
plt.xlabel('Tiempo (s)')
plt.ylabel('Corriente (A) ')
plt.grid(True)
plt.show()

#Percentual error
PeI= abs(Is-Ip)/abs(Is)*100

#Grafica de error acotada a 50 milisegundos
plt.figure(figsize=(8, 6)) 
plt.plot(tp[0:10],PeI[0:10] , label='Percentual error')
#grafica completa
#plt.plot(tp,PeI , label='Percentual error')
plt.legend()
#plt.title('Percentual error betwen simulink and python')
plt.title('Error porcentual entre simulink y python')
plt.grid(True)
plt.xlabel('Tiempo (s)')
plt.ylabel('Error porcentual (%) ')
plt.show()

#Grafica de error acotada a 50 milisegundos
plt.figure(figsize=(8, 6)) 
#plt.plot(tp[0:10],PeI[0:10] , label='Percentual error')
#grafica completa
plt.plot(tp,PeI , label='Percentual error')
plt.legend()
#plt.title('Percentual error betwen simulink and python')
plt.title('Error porcentual entre simulink y python')
plt.grid(True)
plt.xlabel('Tiempo (s)')
plt.ylabel('Error porcentual (%) ')
plt.show()
```

## Las respuestas obtenidas solo difieren significativament al principio de la simulación (durante los 4 primeros milisegundos), pero en general tiende a ser muy similares con el tiempo, así mismo el error porcentual entre estas dos respuestas tiende a ser menor al 1% con el paso del tiempo.
